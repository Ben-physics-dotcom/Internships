```python
column = ['Current [A]','Volt [V]','Power [W]','Ressistent [R]','Source Delay [s]', 'Volt. Lim. [V]']
r_lst_2 = []
for i in range(len(r_list)):
    df = r_list[i]
    df.columns = column
    df = df.drop([0])
    for i in df.columns:
        df[i] = pd.to_numeric(df[i])
    r_lst_2.append(df)
```

```python
r_index = []
for i in range(26):
    x = f'r{i}'
    r_index.append(x)
```

```python
wanted_T = []
for i in range(26):
    if '295K' in r_other[i]:
        x = 295
        wanted_T.append(295)
    if '300K' in r_other[i]:
        wanted_T.append(300)
# wanted_T
```

```python
split_r = []
for i in range(len(r_other)):
    x= r_other[i].split()
    split_r.append(x)
```

```python
I = []
for i in range(len(split_r)):
    x = split_r[i][2].split(sep='_')
    y = x[2].split(sep='(')
    z = y[0].split(sep='I')
    if '0' == z[1]:
        t1 = 0
        I.append(t1)
    else:
        t1 = z[1].split(sep='m')
        I.append(int(t1[0]))
```

```python
t = []
for i in range(len(split_r)):
    x= split_r[i][0].split(sep='_')
    t.append(x[1])
```

```python
file_number = []
for i in range(len(split_r)):
    x= int(split_r[i][1])
    file_number.append(x)
```

```python
T1 = []
T2 = []

for i in range(len(split_r)):
    x= split_r[i][2].split(sep='=')
    y = split_r[i][3].split(sep='=')
    x1 = x[1].split(sep='K')
    y1 = y[1].split(sep='K')
    T1.append(float(x1[0]))
    T2.append(float(y1[0]))
```

```python
# r_index
```

```python
df_r = pd.DataFrame({
    'T1 [K]':T1,
    'Sample T [K]':T2,
    'Wanted T [K]':wanted_T,
    'file number':file_number,
    'Time':t
})
df_r.index = r_index
df_r['I [A]'] = I
df_r.head()
```
